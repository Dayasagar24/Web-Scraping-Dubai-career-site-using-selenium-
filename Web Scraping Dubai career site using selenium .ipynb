```python
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By 
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd 
import time
from selenium.webdriver.support.ui import Select
```

```python
#Dubai careers web site
```

```python
driver=webdriver.Chrome()
driver
```

```python
driver.implicitly_wait(5)
```

```python
driver.get("https://dubaicareers.ae/en/pages/default.aspx")
driver
```

```python
# To click Button
clickbutton = driver.find_element(By.XPATH,'//div[@class="icon fresh-graduate"]')
clickbutton.click()
```

```python
#wait 8 sec for page loading time
import time
time.sleep(8)
```

```python
# To Select dropdawn using select_by_index number
dropdawn = Select(driver.find_element(By.XPATH,'//select[@id="requisitionListInterface.dropListSize"]'))
dropdawn.select_by_index(4)
dropdawn
```

```python
time.sleep(8)
```

```python
# Scraping the job Designations 
Designationx=[]

Designation = driver.find_elements(By.XPATH,'//span[@class="titlelink"]')
for i in Designation:
    name=i.text.strip()
    Designationx.append(name)
Designationx=Designationx[0:30]
print(Designationx)
len(Designationx)
```

```python
# Scraping job category 
job_category=[]

category_tags = driver.find_elements(By.XPATH,"//span[contains(@id,'reqJobField')]")

for tag in category_tags:
    text = tag.text.strip()

    if text and text.lower() != "job category":
        job_category.append(text)
job_category = job_category[:30]    
print(job_category)  
len(job_category)
```

```python
# Scraping job_posting
job_posting=[]

posting_tags = driver.find_elements(By.XPATH,"//span[contains(@id,'reqPostingDate')]")

for i in posting_tags:
    name = i.text.strip()

    if name.lower() != "job posting":
        job_posting.append(name)
job_posting = job_posting[:30]
print(job_posting)
len(job_posting)
```

```python
Employer = []

employer_tag = driver.find_elements(By.XPATH,"//span[contains(@id,'reqOrganization')]")

for tag in employer_tag:
    name = tag.text

    if name.lower() != "employer":
        Employer.append(name)
Employer=Employer[:30]
print(Employer)
len(Employer)
```

```python
df=pd.DataFrame({"Designations":Designationx,"job_category":job_category,"job_posting":job_posting,"Employer":Employer})
df
```

```python
df.to_excel("Dubai careers.xlsx")
```
